```python
# Dependencies and Setup
import json
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
import datetime

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "../Resources/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)

```

## Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#original_cities = cities
#cities = ("new york city", "miami", "gretel city", "los angeles", "sydney", "lima")
#cities

new_cities= []
cloudiness = []
country = []
date = []
humidity = []
temp = []
lat = []
lng = []
wind = []
city_counter = 0
```

```python
#Make API request with test cities
#name url

url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

#f"{url}appid={api_key}&units={units}&q="


for city in cities:
    query_url = url + "appid=" + weather_api_key + "&units=" + units + "&q=" + city
    # Get weather data
    response = requests.get(query_url).json()
    #print(json.dumps(response, indent=4, sort_keys=True))
    
    time.sleep(2)
    
    print(f"Searching for {city} {city_counter + 1}")
    
    city_counter = city_counter + 1
    
    try: 
    
        #append info to lists
        new_cities.append(response['name'])
        cloudiness.append(response['clouds']['all'])
        country.append(response['sys']['country'])
        date.append(response['dt'])
        humidity.append(response['main']['humidity'])
        temp.append(response['main']['temp_max'])
        lat.append(response['coord']['lat'])
        lng.append(response['coord']['lon'])
        wind.append(response['wind']['speed'])
        print(f"added {city_counter}: {city}")
        
        
    except KeyError: 
        print(f"{city} not found.")

```

## Convert Raw Data to DataFrame
* Save as csv

```python
weather_data = {
    "City": new_cities,
    "Cloudiness" : cloudiness,
    "Country" : country,
    "Date" : date,
    "Humidity" : humidity,
    "Temp": temp,
    "Lat" : lat,
    "Lng" : lng,   
    "Wind Speed" : wind
}

print(len(new_cities))
print(len(cloudiness))
print(len(country))
print(len(date)) 
print(len(humidity))
print(len(temp)) 
print(len(lat)) 
print(len(lng))
print(len(wind))

weather_df = pd.DataFrame(weather_data)
weather_df
```

```python
#convert date
date_mdy =datetime.datetime.fromtimestamp(int(date[0])).strftime('%m/%d/%Y')

date_mdy

weather_df["Date"]= date_mdy
weather_df
```

```python
#output df into outputfile

#output_data_file = "output_data/cities.csv"

weather_df.to_csv(output_data_file, index=False)
```

## Linear Regression

### Write linear regression plot function (optional)

```python
#define southern and northern cities

northern = weather_df.loc[weather_df["Lat"] >= 0.0]
northern.reset_index(inplace=True)

northern

southern = weather_df.loc[weather_df["Lat"] < 0.0]
southern.reset_index(inplace=True)

southern
```

```python
def plotLinearRegression(xdata,ydata,xlbl,ylbl,lblpos,ifig):
    (slope, intercept, rvalue, pvalue, stderr) = linregress(xdata, ydata)
    print(f"The r-value is: {rvalue}")
    regress_values = xdata * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

    plt.scatter(xdata,ydata)
    plt.plot(xdata,regress_values,"r-")
    plt.annotate(line_eq,lblpos,fontsize=15,color="red")
    plt.xlabel(xlbl)
    plt.ylabel(ylbl)
    
```

###  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
xlbl = "Lat"
ylbl = "Temp"
lblpos = (50, 60)
plotLinearRegression(northern[xlbl],northern[ylbl],xlbl,ylbl,lblpos,5)
plt.title("Northern Hemisphere - Max Temp vs. Latitude")
plt.xlabel("Latitude")
plt.ylabel("Max Tempurature (F)")


plt.savefig("../Images/NorthLatitudeVsMaxTemp.png")
plt.show
```
